```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import scipy 
import PIL 
import requests

np.random.seed(613)                    # Set the random seed for NumPy's random number generator to 613
x = np.arange(50)                      # Generate an array x that contains numbers from 0 to 49
yl = np.random.randint(0, 100,50)      # two arrays yl and y2 are created, each containing 50 random integers from 0 to 99
y2 = np.random.randint(0, 100,50)

plt.style.use('default')

fig, ax = plt.subplots(figsize=(5, 3)) # This creates a figure & an axis (fig, ax) using plt.subplots() with a width of 5" & height of 3"
ax.plot(x,yl)                          # The plot() function creates a line plot. 2 line plots -  yl & y2 arrays against x on the same axis (ax)
ax.plot(x,y2)                            
fig.show()                             # Displays the plots
```

```python
# Adding a Legend

fig, ax= plt.subplots(figsize=(5, 3)) 
ax.plot(x,yl, label="Person 1")             # label="Person 1" argument adds a label to the first line, which will appear in the plot's legend
ax.plot(x,y2, label="Person 2")             # label="Person 2" argument adds a label to the second line, which will appear in the plot's legend
ax.legend(loc='upper right')                # specify the location of the legend - upper right
fig.show()
```

```python
# Modifying a Legend

fig, ax= plt.subplots(figsize=(5, 3)) 
ax.plot(x,yl, label= "Person 1") 
ax.plot(x,y2, label= "Person 2") 
ax.legend(loc='lower right',
    frameon = True,             # add frame around the legend 
    fontsize = 12,              # change font size
    ncol = 4,                   # specify number of columns
shadow= True)                   # frame with shadow
fig.show()
```

```python
# Moving the Legend Outside of the Plot Area

fig, ax= plt.subplots(figsize=(5, 3)) 
ax.plot(x,yl, label= "Person 1") 
ax.plot(x,y2, label= "Person 2") 
ax.legend(loc='upper left',
          bbox_to_anchor = (1,0.5))     # specify where to anchor the legend on the figure
                                        # Instead of placing the legend inside the plot area, this argument moves the legend outside of the plot
                                        # (1, 0.5) tuple means that the legend box will be anchored at the point (1, 0.5) relative to the plot's axes
                                
fig.show()
```

```python
# Annotation with the text() function

fig, ax=plt.subplots(figsize=(5, 3)) 
ax.scatter(x,yl, label= "Person 1") 
ax.scatter(x,y2, label= "Person 2") 
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!")    # The ax.text() function adds text to the plot at coordinates (10, 95)

fig.show()
```

```python
# Modifying text annotations

ig, ax=plt.subplots(figsize=(5, 3)) 
ax.scatter(x,yl, label= "Person 1") 
ax.scatter(x,y2, label= "Person 2") 
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!", 
        ha='center',	        # alignment of text ha='center': the text will be centered horizontally around the point (10, 95)
        color= 'red',	        # modify font colour 
        size= 20)	        # modify font size
fig.show()
 
```

```python
# Positioning text annotations

fig, ax= plt.subplots() 
ax.axis([0, 10, 0, 10]) # Start by making a simple, empty set of axes (sets the limits for the x and y axes from 0 to 10)

# Then position our annotation with reference to the data, the axes, or the figure (referencing axes/figure dimensions)
ax.text(1, 5,  ". Data:(l, 5 )" ,	                transform = ax.transData)   # adds a text annotation at the coordinates (1, 5) in data space
ax.text(0.5, 0.1,   ". Axes:(0.5, 0.1)" ,	        transform=ax.transAxes)     # adds a text annotation at the coordinates (0.5, 0.1) in axes space
ax.text(0.2, 0.2,    ". Figure: (0.2, 0.2 )" ,	    transform=fig.transFigure)  # adds a text annotation at the coordinates (0.2, 0.2) in figure space
 
```

```python
# Annotating with Arrows
    # Use the annotate(} function to annotate with arrows and text (e.g. to focus at a specific point)
    # Need to specify the location to point at & the location where we want our text

fig, ax=	plt.subplots(figsize=(5, 3)) 
ax.scatter(x,yl, label= "Person 1"  ) 
ax.scatter(x,y2, label= "Person 2" ) 
ax.legend(loc='lower right')


ax.annotate('This is important!', xy=(10, 95), xytext=(20, 94), 
            arrowprops=dict(facecolor='red', shrink=0.01))

fig.show()

# xy=(10, 95) is the point being annotated. It specifies the coordinates on the plot where the annotation point to. 
# This is the start point of the arrow. In this case, the arrow will point to the coordinates (10, 95) on the plot.

# xytext=(20, 94) is the location of the annotation text. It specifies where the text label itself will be placed, which is at the coordinates (20, 94) on the plot. 
# This is where the actual label "This is important!" will appear
```

```python
# Modifying Arrows
        # Use the arrowprops argument to adjust the arrow style & colour

ax.annotate('This is important!',
        xy=(10, 95), xytext=(20, 94), 
        arrowprops = dict(arrowstyle = "wedge",     # Use the arrowprops argument to adjust the arrow style & colour   
                            color= "hotpink"))

fig.show()


```

```python
#   Limiting tick marks


fig, ax= plt.subplots(figsize=(5, 3)) 
ax.scatter(x,yl, label= "Person 1") 
ax.scatter(x,y2, label= "Person 2") 
ax.legend(loc='lower right') 
ax.xaxis.set_major_locator(plt.MaxNLocator(3))  # set maximum number of ticks (in this case, 3) using MaxNLocator
```

```python
# Data
np.random.seed(613)
xl = np.arange(50)
yl = np.random.randint(0, 75, 50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])

# Plot
fig, (axl, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(7, 3))

# Scatter plot
axl.scatter(xl, yl)

# Bar plot with categorical x-axis
ax2.bar(x2, y2)

# Display the plot
fig.show()
```

```python
# Customizing Subplots

fig, (ax1, ax2) = plt.subplots (ncols=2, nrows=1, figsize=(10,3))

ax1.scatter(xl, yl, color="violet")
ax2.bar(x2, y2, color='red')
ax1.set_title('My Scatterplot')
ax2.set_title('My Bar Chart')

ax1.annotate('This is fun!', xy=(10, 60), xytext=(20, 60), 
            arrowprops=dict(facecolor='red', shrink=0.01))

fig.show
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],          # The list [['ax1', 'ax3'], ['ax2', 'ax3']] defines the grid of subplots:
                                    ['ax2','ax3']],             # 'ax1' is placed in the top-left, 'ax3' is placed in both the top-right and bottom-right
                                    figsize = (7,4))            # 'ax2' is placed in the bottom-left

for label, ax in someaxes.items():            # Loops through the items in the dictionary, where label will be 'ax1', 'ax2', or 'ax3', and ax will be the respective axis object i.e. ax1, ax2, or ax3
    ax.text(0.5, 0.5, f'This is {label!r}',   # f'This is {label!r}' is an f-string that dynamically creates a string. It uses {label!r} to include the label of the subplot (e.g., 'ax1', 'ax2', or 'ax3'
            fontsize=14,                        
            ha = 'center',                    # Add text to the subplot horizontally & centred aligned in each subplot
            transform=ax.transAxes)           # The text() function places the annotation at (0.5, 0.5) relative to the axes' bounding box
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],              
                                    ['ax2','ax3']],             
                                    figsize = (7,4))            

someaxes["ax1"].scatter(xl,yl)                                  # Scatter plot on ax1.
someaxes["ax2"].bar(x2,y2)                                      # Bar plot on ax2.
someaxes["ax3"].plot(xl,yl)                                     # Line plot on ax3.

plt.show()
```

```python
#  LAYOUTS

fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],              # The list [['ax1', 'ax3'], ['ax2', 'ax3']] defines the grid of subplots:
                                    ['ax2','ax3']],             # 'ax1' is placed in the top-left, 'ax3' is placed in both the top-right and bottom-right 
                                    figsize = (7,4))            # 'ax2' is placed in the bottom-left

someaxes["ax1"].scatter(xl,yl)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(xl,yl)

someaxes["ax1"].set_xlabel('A Big Label', fontsize = 18)
someaxes["ax2"].set_xlabel('Another Big Label', fontsize = 18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 14)

```

```python
# Constrained Layout

fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2','ax3']],
                                    figsize = (7,4),
                                    layout = "constrained") 

# layout="constrained" ensures that the spacing between subplots is automatically adjusted so that the labels, titles, and other elements do not overlap

someaxes["ax1"].scatter(xl,yl)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(xl,yl)

someaxes["ax1"].set_xlabel('A Big Label', fontsize = 18)
someaxes["ax2"].set_xlabel('Another Big Label', fontsize = 18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 14)
```

```python
# Tight Layout

fig, someaxes = plt.subplot_mosaic([['ax1','ax3'],
                                    ['ax2','ax3']],
                                    figsize = (7,4),
                                    layout = "tight")

# "tight" layout automatically adjusts the spacing between subplots in the figure to reduce any excess empty space around & between the subplots. 
# It essentially "tightens" the layout, making the subplots more compact.

someaxes["ax1"].scatter(xl,yl)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(xl,yl)

someaxes["ax1"].set_xlabel('A Big Label', fontsize = 18)
someaxes["ax2"].set_xlabel('Another Big Label', fontsize = 18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious', fontsize = 14)
```

```python
# Multiple viz on one axes 

x = np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y1 = np.array([110,180,240,99,220])
y2 = np.array([170,100,90,120,50])

# define our figure and axes 
fig, ax = plt.subplots(figsize=(7,3))

# call both bar & plot to the same axes
ax.bar(x, y1, color = "indigo")
ax.plot(x, y2, color = "red")
```

```python
# Add Error Information

# Calculate std dev of data
y2_sd = np.std(y2)

# Plot the line
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2,color = 'red')

# Add error bar & customizing error bar appearance
ax.errorbar(x,y2,
            yerr=y2_sd,         # plotting vertical error bars
            fmt= "none",        # not plotting the actual data points, only the error
            ecolor = "indigo",  # color of the error bars
            capsize = 6,        # length of the cap
            capthick = 6)       # thickness of the cap
```
